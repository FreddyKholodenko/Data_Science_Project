```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time

def load_html_page(url):
    userAgent = {'User-agent': '*'}
    resp = requests.get(url, headers=userAgent)
    return BeautifulSoup(resp.content,"html.parser")
```

```python
soup = load_html_page("https://www.metacritic.com/browse/games/score/metascore/all/all/filtered")
game_name = []
game_release_year =[]
game_critics_amount=[]
game_critics_score=[]
game_user_score=[]
game_genre = []
game_genre_1 = []
game_genre_2 = []
game_genre_3 = []
game_genre_4 = []
game_genre_5 = []
game_genre_6 = []
game_genre_7 = []
game_genre_8 = []
game_genre_9 = []
game_genre_10 = []
game_genre_11 = []
game_genre_12 = []
df=pd.DataFrame()
next_page = soup.find("span",attrs={"class":"flipper next"}).a['href']
while next_page != None:
    try:
        game_info_list=soup.findAll("td",attrs={"class":"clamp-summary-wrap"})
    except:
        time.sleep(600)
        game_info_list=soup.findAll("td",attrs={"class":"clamp-summary-wrap"})
    for td in game_info_list:
        sp = td.find("div",attrs={"class":"clamp-details"}).select("span")
        release_date_string= str(sp[2].text)
        release_year_string=release_date_string[len(release_date_string)-4]+release_date_string[len(release_date_string)-3]+release_date_string[len(release_date_string)-2]+release_date_string[len(release_date_string)-1]
        if int(release_year_string)>=2007:
            if sp[1].text != "iOS":
                try:
                    new_soup=load_html_page("https://www.metacritic.com/"+td.find("a",attrs={"class":"title"})['href'])
                except:
                    time.sleep(600)
                    new_soup=load_html_page("https://www.metacritic.com/"+td.find("a",attrs={"class":"title"})['href'])
                try:    
                    game_name.append(new_soup.find("div",attrs={"class":"product_title"}).a.text.replace("\n",""))
                except:
                     game_name.append(None)
                game_release_year.append(release_year_string)
                try:
                    game_critics_amount.append(new_soup.find("span",attrs={"class":"count"}).a.span.text.replace("\n",""))
                except:
                    game_critics_amount.append(None)
                try:
                    game_critics_score.append(new_soup.find("div",attrs={"class":"metascore_w xlarge game positive"}).text)
                except:
                    try:
                        game_critics_score.append(new_soup.find("div",attrs={"class":"metascore_w xlarge game mixed"}).text) 
                    except:
                        try:
                            game_critics_score.append(new_soup.find("div",attrs={"class":"metascore_w xlarge game negative"}).text)
                        except:
                            game_critics_score.append(None)
                try:
                    game_user_score.append(new_soup.find("div",attrs={"class":"metascore_w user large game positive"}).text)
                except:
                    try:
                        game_user_score.append(new_soup.find("div",attrs={"class":"metascore_w user large game mixed"}).text) 
                    except:
                        try:
                            game_user_score.append(new_soup.find("div",attrs={"class":"metascore_w user large game negative"}).text)
                        except:
                            game_user_score.append(None)
                try:
                    game_genre_span = new_soup.find("li",attrs={"class":"summary_detail product_genre"}).findAll("span",attrs={"class":"data"})
                except:
                    game_genre_span=[]
                    game_genre = []
                for span in game_genre_span:
                    game_genre.append(span.text)
                for i in range(len(game_genre),12):
                    game_genre.append(None)
                game_genre_1.append(game_genre[0])
                game_genre_2.append(game_genre[1])
                game_genre_3.append(game_genre[2])
                game_genre_4.append(game_genre[3])
                game_genre_5.append(game_genre[4])
                game_genre_6.append(game_genre[5])
                game_genre_7.append(game_genre[6])
                game_genre_8.append(game_genre[7])
                game_genre_9.append(game_genre[8])
                game_genre_10.append(game_genre[9])
                game_genre_11.append(game_genre[10])
                game_genre_12.append(game_genre[11])
                game_genre=[]
                time.sleep(5)
    try:
        soup = load_html_page("https://www.metacritic.com/"+next_page)
    except:
        time.sleep(600)
        soup = load_html_page("https://www.metacritic.com/"+next_page)
    if soup.find("span",attrs={"class":"flipper next"}).a == None:
        next_page = None 
    else:
        next_page=soup.find("span",attrs={"class":"flipper next"}).a['href']
        if "browse" not in next_page:
            next_page = "browse"+next_page
    df_1=pd.DataFrame({"game_name":game_name,"game_release_year":game_release_year,"game_critics_amount":game_critics_amount,"game_critics_score":game_critics_score,
                 "game_user_score":game_user_score,"game_genre_1":game_genre_1,"game_genre_2":game_genre_2,
                 "game_genre_3":game_genre_3,"game_genre_4":game_genre_4,"game_genre_5":game_genre_5,
                 "game_genre_6":game_genre_6,"game_genre_7":game_genre_7,"game_genre_8":game_genre_8,"game_genre_9":game_genre_9,
                 "game_genre_10":game_genre_10,"game_genre_11":game_genre_11,"game_genre_12":game_genre_12})
    df=df.append(df_1)
    game_name = []
    game_release_year =[]
    game_critics_amount=[]
    game_critics_score=[]
    game_user_score=[]
    game_genre = []
    game_genre_1 = []
    game_genre_2 = []
    game_genre_3 = []
    game_genre_4 = []
    game_genre_5 = []
    game_genre_6 = []
    game_genre_7 = []
    game_genre_8 = []
    game_genre_9 = []
    game_genre_10 = []
    game_genre_11 = []
    game_genre_12 = []
    time.sleep(5)
df
```
